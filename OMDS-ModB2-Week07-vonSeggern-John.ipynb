# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
# Recreating demonstrating Effectiveness
from matplotlib.ticker import PercentFormatter

df = pd.DataFrame({
    'data':[12.9,12.3,11.6,11.6,11.0,10.3,10.0,8.8,8.6,8.2], 
    'labels':['Engine power is less than expected','Tires make excessive noise while driving','Engine makes abnormal/excessive noise','Seat material concerns','Excessive wind noise','Hesitation or delay when shifting','Bluetooth system has poor sound quality','Steering ststem/wheel has too much play','Bluetooth system is difficult to use','Front seat audio/entertainment/navigation controls']
    })

# Text to add nest to the chart
first = ("Comments indicate that\nnoisy tires issues are\nmost apparent in the rain.")
second = ("Complaints about engine\nnoise commonly cited\nafter the car had not\nbeen driven for a while.")
third = ("Excessive wind noise is\nnoted primarily in freeway\ndriving at high speeds.")

# bold_text = '\x1B[1m' + 'This text is bold.' + '\x1B[0m'
# red_font = "\033[31mThis text is red.\033[0m"

title = ('Of the top design concerns, three are noise-related')
subtitle = ('Top 10 design concerns')

# Choose which category to highlight
highlight_category = ['Tires make excessive noise while driving','Engine makes abnormal/excessive noise','Excessive wind noise']
colors = ['red' if cat in highlight_category else 'grey' for cat in df["labels"]]

# Plot the data and labels
plt.figure(figsize=(8, 5))
ax = sns.barplot(x='data', y='labels', data=df, palette=colors, legend=False)

for container in ax.containers:
    ax.bar_label(container)

# Move x-axis to the top
ax.xaxis.set_label_position('top')
plt.gca().set_xticks([])  

# Removes spines from the chart
sns.despine(left=True, bottom=True)
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(False)

# Add labels and title
plt.xlabel("concerns per 1,000")
plt.title(title, x=-.61, pad=75, loc='left', color='red')

# add subtitle to the chart
ax.text(x=-.35, y=.95, s=subtitle, fontsize=20, ha='left', transform=plt.gcf().transFigure)

# Adding the demo text next to "Demonstration of results"
ax.text(15, df[df["labels"] == "Engine power is less than expected"].index[0], 
        first, fontsize=10, verticalalignment='top', color='black')

ax.text(15, df[df["labels"] == "Engine makes abnormal/excessive noise"].index[0], 
        second, fontsize=10, verticalalignment='top', color='black')

ax.text(15, df[df["labels"] == "Excessive wind noise"].index[0], 
        third, fontsize=10, verticalalignment='top', color='black')

# Removing y label
plt.ylabel(None)
plt.show()
```
