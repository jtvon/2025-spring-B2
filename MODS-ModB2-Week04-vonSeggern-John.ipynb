# Histogram Variants

```python
# Histograms for all numerical features in the matches dataset
import matplotlib.pyplot as plt
import seaborn as sns

df_match.hist(figsize=(20,15))
None
```

### Comparing winner vs loser stats

```python
# Serve Points
df_svpt = df_match[df_match['w_svpt'] < 225]
sns.kdeplot(df_svpt[['w_svpt','l_svpt']], )
plt.title('Serve points won by the individual')
plt.xlabel('Points won')
plt.ylabel('Count')
None
```

The winners of a tennis match tend to have lower average points won compared to the losers of a tennis match. It seems as the match progresses, the losers have an overall higher average of points won.

```python
# 1st serve in
df_1stin = df_match[df_match['w_1stIn'] < 110]
plt.hist(df_1stin[['w_1stIn','l_1stIn']], bins=25, label=['winner','loser'])
plt.legend()
plt.title('1st Serve in by the individual')
plt.xlabel('Serves In')
plt.ylabel('Count')
None
```

In the above graphic, winners tend to have a higher count in the smaller ranges of serves in. This could be due to winners of tennis matches facing less opposition on their serve, but are good at creating opposition on their opponents serve, hence why losers get more first serves in as the match length increases.

```python
# Break Points Faced
df_bp = df_match[df_match['w_bpFaced'] < 25]
plt.hist(df_bp[['w_bpFaced','l_bpFaced']], bins=25, label=['winner','loser'])
plt.legend()
plt.title('Break Points faced by the individual')
plt.xlabel('Break Points')
plt.ylabel('Count')
None
```

In this chart, we see that winners are less likely to face break points in a match. This follows that pattern we have seen with the other charts, where short matches favor the stats of the winner, and the longer matches favor the stats of the loser.

```python
# Break Points Saved
df_bpsv = df_match[df_match['w_bpSaved'] < 25]
plt.hist(df_bpsv[['w_bpSaved','l_bpSaved']], bins=27, label=['winner','loser'])
plt.legend()
plt.title('Break Points saved by the individual')
plt.xlabel('Break Points')
plt.ylabel('Count')
None
```

For the last graphic, we have number of break points saved by the individual. There is a spike on the left side of the graph indicating that winners are more likely to save a fewer amount of break points, but once the number of break points gets to 4 or more, the losers maintain the edge over the winners. This chart can be interpreted one of two ways, either the winner is not as good at saving break points as the loser, or the winner is less likely to face break points altogether. I'm more inclined to think the latter.
