Reproduce any graph of your choice in p. 52-68 of the Storytelling With Data book as best you can.  (The second half of chapter two).  You do not have to get the exact data values right, just the overall look and feel.

```python
df_rank = df_match.groupby('surface')[['w_bpFaced', 'w_svpt', 'winner_ht']].mean()
df_rank = df_rank.dropna()
df_rank.head()
```

```python
# Multi-horizontal bar chart grouped by surface
categories = ['Carpet', 'Clay', 'Grass', 'Hard']

# Position of the bars on the y-axis
y_pos = np.arange(len(categories))

# Width of the bars
bar_width = 0.2

# Create the plot
plt.figure(figsize=(10,6))

# Plot the bars for each series
plt.barh(y_pos + bar_width, df_rank['winner_ht'], bar_width, label='Height', color='lightgreen')
plt.barh(y_pos, df_rank['w_svpt'], bar_width, label='Service Points', color='lightcoral')
plt.barh(y_pos - bar_width, df_rank['w_bpFaced'], bar_width, label='Break Points Faced', color='skyblue')

# Set the y-axis labels
plt.yticks(y_pos, categories)

# Add labels and title
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Multiseries Horizontal Bar Chart of Averages')

# Add legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
None
```
